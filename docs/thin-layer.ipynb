## Zero thickness layer


```python
s = iadpython.Sample(n=1.5, quad_pts=8)
s.update_quadrature()

r, t = iadpython.zero_layer(s)
s.wrmatrix(r)
s.wrmatrix(t)
```

## IGI start method

```python
s = iadpython.Sample(a=1, b=100, g=0)

r, t = iadpython.igi(s)
s.wrmatrix(r)
s.wrmatrix(t)
```

```python
s = iadpython.Sample(a=1, b=100, g=0.9, quad_pts=4)

r, t = iadpython.igi(s)
s.wrmatrix(r)
s.wrmatrix(t)
```
